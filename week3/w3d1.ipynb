```python
def rng(seed=1):
    def rand():
        nonlocal seed
        seed = (7 * seed) % 101
        #print("Seed: {}".format(seed))
        return (seed - 1) % 10 + 1

    return rand

rand = rng(135135)
```

```python
hist = Histogram(nums, bins=10, tools=False,
                 title="rng", ylabel="frequency", xlabel="distributions")
show(hist)
```

```python
def bsd_rng(seed=1):
    def rand():
        nonlocal seed
        seed = (1103515245 * seed + 12345) & 0x7fffffff
        return seed / 2147483647
    return rand
```

```python
rand = bsd_rng()
nums = [rand() for _ in range(10000)]
```

```python
nums
```

```python
hist = Histogram(nums, bins=30, tools=False, title="bsd_rng", ylabel="frequency", xlabel="distributions")
show(hist)
```

```python
nums = [random.random() for _ in range(10000)]
hist = Histogram(nums, bins=30, tools=False,
                 title="random.random", ylabel="frequency", xlabel="distributions")
show(hist)
```

```python
nums = [random.randint(1, 100) for _ in range(10000)]
hist = Histogram(nums, bins=30, tools=False,
                 title="random.randint", ylabel="frequency", xlabel="distributions")
show(hist)
```

```python
choices = range(1, 101)
nums = [random.choice(choices) for _ in range(10000)]
hist = Histogram(nums, bins=30, tools=False,
                 title="random.choice", ylabel="frequency", xlabel="distributions")
show(hist)
```

```python
nums = [random.triangular(0, 1) for _ in range(10000)]
hist = Histogram(nums, bins=30, tools=False,
                 title="random.triangular", ylabel="frequency", xlabel="distributions")
show(hist)
```

```python
nums = [random.gauss(0.5, 0.5) for _ in range(10000)]
hist = Histogram(nums, bins=30, tools=False,
                 title="random.gauss", ylabel="frequency", xlabel="distributions")
show(hist)
```

```python
nums = [random.expovariate(2) for _ in range(10000)]
hist = Histogram(nums, bins=30, tools=False,
                 title="random.expovariate", ylabel="frequency", xlabel="distributions")
show(hist)
```
